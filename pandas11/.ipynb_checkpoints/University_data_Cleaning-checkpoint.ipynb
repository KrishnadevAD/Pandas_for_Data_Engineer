```python
import numpy as np 
import pandas as pd 
import os
universityTown=[]
filepath=os.path.join(os.getcwd(),r"C:\Users\DELL\Desktop\data science DWIT\day11\university_towns.txt")
fileOpen=open(filepath,"r")
fileRead=fileOpen.readlines()
```

```python
for lines in fileRead:
    print(lines)
```

```python
for lines in fileRead:
    if '[edit]' in lines:
        print(lines)
    print(lines)
```

```python
for lines in fileRead:
    if '[edit]' in lines:
        state = lines
    else:
        universityTown.append((state,lines))
print(universityTown)
```

```python
townDf=pd.DataFrame(universityTown,columns=["State","Region"])
print(townDf)
```

```python
def get_city(items):
    if '(' in items:
        return items[:items.find( '()')]
    elif '[' in items:
        return items[:items.find('[')]
    else:
        return items
```
